12. kilometer - How many kilometers the car has driven.
13. monthOfRegistration - The month in which which year the car was first registered.
14. fuelType - What type of fuel the car uses.
15. brand - The brand of the car.
16. notRepairedDamage - If the car has a damage which is not yet repaired.
17. dateCreated - The date on which the eBay listing was created.
18. nrOfPictures - The number of pictures in the ad.
19. postalCode - The postal code for the location of the vehicle.

The registration year as described has some years that doesnt exist. Years like 9999,8888, which are many years into the future and other years like 1000 when cars doesnt exist. We will be removing such year from our data. Because a car can't be first registered after the listing was seen, any vehicle with a registration year above 2016 is definitely inaccurate. Determining the earliest valid year is more difficult. Realistically, it could be somewhere in the first few decades of the 1900s. so we will be working with years between 1900 to 2016.

```python
autos = autos[autos["registration_year"].between(1900, 2016)]
autos.describe()
```

```python
autos["brand"].value_counts(normalize = True)
```

There are 40 different types of brand in this data. However, since we will be working on brands that have the highest number of ads on this data. There are alot of brands that doesnt have a significant percentage on this listings so we will be limiting our analysis withing brands that have over 2% of the listing

```python
brand = autos["brand"].value_counts(normalize = True)
brand = brand[brand > 0.02]
brand = brand.index
brand
```

```python
mean_price = {}
for values in brand:
    name = autos[autos["brand"] == values]
    mean = name["price"].mean()
    mean_price[values] = int(mean)
mean_price
```


Of the top 9 brands, the price gaps areobvious:

1. Audi, BMW and Mercedes Benz are more expensive
2. Ford, peugeot,fiat, renault and Opel are less expensive
3. Volkswagen is in between
This translates that Volkwagen which has the highest number of brand of cars in this data has the price in betwee which is not too expensive and not too cheap. This is the reason why Volkwagen is the highest

# Exploring Milleage

```python
mean_milleage = {}
for values in brand:
    milleage = autos[autos["brand"] == values]
    mean = milleage["odometer_km"].mean()
    mean_milleage[values] = int(mean)
mean_milleage
```

```python
for_price = pd.Series(mean_price).sort_index(ascending = False)
for_milleage = pd.Series(mean_milleage).sort_index(ascending = False)
final = pd.DataFrame(for_price, columns= ['Price'])
final["Milleage"] = for_milleage
final
```

Expensive Cars have higher milleage than the less expensive cars. However the range at which they vary by mileage is not too much compares to how the prices vary
